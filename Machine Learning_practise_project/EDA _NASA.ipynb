```python
import requests
import pandas as pd

# Your NASA API key
api_key = "E0L8wxjNK1lEKlQc1h3jVFpOj516Zd2gNydb5Q1Q"

# NASA API endpoint for Astronomy Picture of the Day (APOD)
url = "https://api.nasa.gov/planetary/apod"

# Define parameters, including your API key and optional date range
params = {
    "api_key": api_key,
    "start_date": "2024-10-01",  # optional: get data starting from this date
    "end_date": "2024-10-21"  # optional: get data up to this date
}

# Make the request
response = requests.get(url, params=params)

# Check if the request was successful
if response.status_code == 200:
    # Parse the response JSON into a list of dictionaries
    data = response.json()

    # Convert the JSON data to a pandas DataFrame
    df = pd.DataFrame(data)

    # Save the DataFrame to a CSV file
    df.to_csv('nasa_apod_data.csv', index=False)

    print("Data saved successfully to nasa_apod_data.csv")
else:
    print("Error:", response.status_code)
```
