````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.model_selection import StratifiedKFold, train_test_split, GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, log_loss
import warnings
warnings.filterwarnings('ignore')
````

**a.Do the hot encoding / dummying of all the non-numeric variables**

**i.SVC (kernel=’linear’)**

````python
resturant = pd.read_csv('/content/restaurant_customer_satisfaction.csv', index_col=0)
X = resturant.drop('HighSatisfaction', axis=1)
y = resturant['HighSatisfaction']
````

````python
ohe = OneHotEncoder(sparse_output = False, drop='first', handle_unknown='ignore')
ct = make_column_transformer((ohe, make_column_selector(dtype_include=object)), remainder='passthrough')
````

````python
# '1. Classification Scoring Metrics
# accuracy: Proportion of correctly classified samples.
# precision: Measures the accuracy of the positive predictions (true positives / predicted positives).
# recall: Measures the ability to find all positive samples (true positives / actual positives).
# f1: The harmonic mean of precision and recall; useful for imbalanced datasets.
# roc_auc: Area under the ROC curve; best for binary classification to see overall model performance.
# log_loss: Negative log likelihood of the classifier (lower is better); penalizes overconfident wrong predictions.'
````

````python
# 2. Regression Scoring Metrics
# neg_mean_squared_error: Negative Mean Squared Error (MSE); lower MSE is better, but we use the negative value since higher scores indicate better performance in RandomizedSearchCV.
# neg_mean_absolute_error: Negative Mean Absolute Error (MAE); focuses on absolute error.
# r2: R-squared value (1.0 is best, 0 means no fit, negative values mean the model performs worse than a constant).
# neg_mean_squared_log_error: Negative Mean Squared Log Error; works best when the target values are skewed.
# explained_variance: Measures the proportion of variance explained by the model.
````

````python
# 3. Multi-Class Classification Metrics
# When dealing with multi-class classification, you can specify the average type for metrics like f1, precision, and recall:

# f1_macro: Average of F1 scores across classes, treating each class equally.
# f1_micro: Aggregate F1 score by counting total true positives, false negatives, and false positives.
````

````python
# 4. Custom Scoring Functions
# You can create a custom scoring function using make_scorer from sklearn.metrics. For example, if you want to use a custom metric that isn't available by default in scikit-learn.
````

````python
svm = SVC(random_state=24,probability=True)
#log = LogisticRegression(random_state=24,  max_iter= 100)
kfold  = StratifiedKFold(n_splits=5, shuffle=True, random_state=24)
pipe = Pipeline([('CT', ct), ('SVM', svm)])
params = {'SVM__kernel': ['linear'],
          'SVM__decision_function_shape':['ovo', 'ovr'],
          'SVM__degree': np.arange(3, 10),
          'SVM__max_iter': np.arange(5),
          'SVM__gamma': ['scale', 'auto']}
gcv = GridSearchCV(pipe, param_grid=params, cv=kfold, scoring='neg_log_los', verbose=1)
gcv.fit(X, y)
````

````python
gcv.best_params_, gcv.best_score_
````

**ii.SVC (kernel=’rbf’)**

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, log_loss
````

````python
resturant = pd.read_csv('/content/restaurant_customer_satisfaction.csv', index_col=0)
X = resturant.drop('HighSatisfaction', axis=1)
y = resturant['HighSatisfaction']
````

````python
ohe = OneHotEncoder(sparse_output = False, drop='first', handle_unknown='ignore')
ct = make_column_transformer((ohe, make_column_selector(dtype_include=object)), remainder='passthrough')
````

````python
svm = SVC(random_state=24,probability=True)
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=24)
pipe = Pipeline([('CT', ct), ('SVM', svm)])
params = {'SVM__kernel': ['rbf'],
          'SVM__decision_function_shape':['ovo', 'ovr'],
          'SVM__degree': np.arange(3, 10),
          'SVM__max_iter': np.arange(5),
          'SVM__gamma': ['scale', 'auto']}
gcv = GridSearchCV(pipe, param_grid=params, cv=kfold, scoring='neg_log_loss', verbose=1)
gcv.fit(X, y)
````

````python
gcv.best_params_, gcv.best_score_
````

**iv.Random Forest Classifier**

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, log_loss
````

````python
resturant = pd.read_csv('/content/restaurant_customer_satisfaction.csv', index_col=0)
X = resturant.drop('HighSatisfaction', axis=1)
y = resturant['HighSatisfaction']
````

````python
rfc = RandomForestClassifier(random_state=24)
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=24)
pipe = Pipeline([('CT', ct), ('RFC', rfc)])
params = {'RFC__n_estimators': [100, 200, 300],
          'RFC__criterion': ['gini', 'entropy'],
          'RFC__max_depth': [1, 3, 5]}
rgcv = RandomizedSearchCV(pipe, param_distributions=params, cv=kfold, scoring='neg_log_loss', verbose=1)
rgcv.fit(X, y)
````

````python
rgcv.best_params_, rgcv.best_score_
````
