```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
import os
os.chdir(r'C:\Training\Academy\Statistics (Python)\Datasets')
```

```python
groceries = []
with open("groceries.csv","r") as f:groceries = f.read()
groceries = groceries.split("\n")

groceries_list = []
for i in groceries:
    groceries_list.append(i.split(","))
groceries_list[:3]
```

```python
te = TransactionEncoder().set_output(transform='pandas')
te_ary = te.fit_transform(groceries_list)
itemsets = apriori(te_ary, min_support=0.01, use_colnames=True)
rules = association_rules(itemsets, metric='lift', min_threshold=1,
                          num_itemsets=len(te_ary),
                          return_metrics=[ 'support', 'confidence', 'lift'])
rules.sort_values('confidence', ascending=False)
```
